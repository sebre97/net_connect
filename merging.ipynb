```python
import pandas as pd
import math
import time 
import gc 
```

```python
df = pd.read_csv('Daten_Raster_Punkte\Ppoints_rtr' + str(1000000) + '.csv',index_col=0)
```

```python
list_tmp = [[]] * len(df)

for i in range(len(df)):
    if df['IndexListe'][i] != "[]":
        tmp = df['IndexListe'][i]
        tmp = tmp.replace("[","").replace("]","")
        tmp = tmp.split(",")
        tmp = [int(x) for x in tmp]
        list_tmp[i] = tmp
     
df['IndexListe'] = list_tmp
     
```

```python
for j in range(1005000, 1875000, 5000):#1875000
    starttime=time.time()
    df_tmp = pd.read_csv('Daten_Raster_Punkte\Ppoints_rtr' + str(j) + '.csv')
    for i in range(len(df)):
        if df_tmp['IndexListe'][i] != "[]":
            tmp = df_tmp['IndexListe'][i]
            tmp = tmp.replace("[","").replace("]","")
            tmp = tmp.split(",")
            tmp = [int(x) for x in tmp]
            df.loc[i, 'IndexListe'].extend(tmp)
            del tmp

            
    print(str(j)+" printed")
    print(starttime-time.time())        
    del df_tmp
    gc.collect()
```

```python
for i in range(0,len(df["IndexListe"])):
    
    df["IndexListe"][i] = list(set(df["IndexListe"][i])) 
```
