```python
for i in range(0,len(AUT_map_Merged)):
    summe1 = 0 
    summe2 = 0
    summe3 = 0
    summe4 = 0
    summe5 = 0
    summe6 = 0
    summe7 = 0
    summe8 = 0
    summe9  = 0
    
    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0
    count5 = 0
    count6 = 0
    count7 = 0
    count8 = 0
    count9 = 0
    
    liste = AUT_map_Merged["IndexListe"][i].strip('][').split(', ')
    for j in liste:
        if j != "":
            if math.isnan(df_clc.loc[int(j),"download_kbit"]) == False:
                summe1  = summe1 + df_clc.loc[int(j),"download_kbit"]
                count1 = count1 + 1
            if math.isnan(df_clc.loc[int(j),"upload_kbit"]) == False:
                summe2  = summe2 + df_clc.loc[int(j),"upload_kbit"]
                count2 = count2 + 1         
            if math.isnan(df_clc.loc[int(j),"ping_ms"]) == False:
                summe3  = summe3 + df_clc.loc[int(j),"ping_ms"]
                count3 = count3 + 1
            if math.isnan(df_clc.loc[int(j),"lte_rsrp"]) == False:
                summe4  = summe4 + df_clc.loc[int(j),"lte_rsrp"]
                count4 = count4 + 1
            if math.isnan(df_clc.loc[int(j),"lte_rsrq"]) == False:
                summe5  = summe5 + df_clc.loc[int(j),"lte_rsrq"]
                count5 = count5 + 1
            if math.isnan(df_clc.loc[int(j),"ndt_download_kbit"]) == False:
                summe6  = summe6 + df_clc.loc[int(j),"ndt_download_kbit"]
                count6 = count6 + 1
            if math.isnan(df_clc.loc[int(j),"ndt_upload_kbit"]) == False:
                summe7  = summe7 + df_clc.loc[int(j),"ndt_upload_kbit"]
                count7 = count7 + 1
            if math.isnan(df_clc.loc[int(j),"signal_strength"]) == False:
                summe8  = summe8 + df_clc.loc[int(j),"signal_strength"]
                count8 = count8 + 1
            if math.isnan(df_clc.loc[int(j),"dtm_level"]) == False:
                summe9  = summe9 + df_clc.loc[int(j),"dtm_level"]
                count9 = count9 + 1

            
    if count1 == 0:
         AUT_map_Merged.loc[i, "AverageDOWNLOAD"] = float('nan')
    else:      
        AUT_map_Merged.loc[i, "AverageDOWNLOAD"] = summe1/max(count1,1) 
        
    if count2 == 0:
        AUT_map_Merged.loc[i, "AverageUPLOAD"] = float('nan')
    else:  
        AUT_map_Merged.loc[i, "AverageUPLOAD"] = summe2/max(count2,1)
        
    if count3 == 0:
        AUT_map_Merged.loc[i, "Averageping_ms"] = float('nan')
    else:
        AUT_map_Merged.loc[i, "Averageping_ms"] = summe3/max(count3,1)
        
    if count4 == 0:
        AUT_map_Merged.loc[i, "Averagelte_rsrp"] = float('nan')
    else:
        AUT_map_Merged.loc[i, "Averagelte_rsrp"] = summe4/max(count4,1)
        
    if count5 == 0:
        AUT_map_Merged.loc[i, "Averagelte_rsrq"] = float('nan') 
    else:
        AUT_map_Merged.loc[i, "Averagelte_rsrq"] = summe5/max(count5,1) 
        
    if count6 == 0:
        AUT_map_Merged.loc[i, "Averagendt_download_kbit"] = float('nan')  
    else:
         AUT_map_Merged.loc[i, "Averagendt_download_kbit"] = summe6/max(count6,1) 
            
    if count7 == 0:
        AUT_map_Merged.loc[i, "Averagendt_upload_kbit"] = float('nan') 
    else:
        AUT_map_Merged.loc[i, "Averagendt_upload_kbit"] = summe7/max(count7,1)
        
    if count8 == 0:
        AUT_map_Merged.loc[i, "Averagesignal_strength"] = float('nan') 
    else:
        AUT_map_Merged.loc[i, "Averagesignal_strength"] = summe8/max(count8,1)

    if count9 == 0:
        AUT_map_Merged.loc[i, "Averagedtm_level"] = float('nan') 
    else:
         AUT_map_Merged.loc[i, "Averagedtm_level"] = summe9/max(count9,1)
    

        #df_clc["IndexListe"][j]
        
    
```
